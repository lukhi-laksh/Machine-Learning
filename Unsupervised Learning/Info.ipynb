# What is UnSupervised Machine Learning

Unsupervised learning is a type of machine learning that learns from unlabeled data.

This means that the data does not have any pre-existing labels or categories.

The goal of unsupervised learning is to discover patterns and relationships in the data without any explicit guidance.

```python
# --- Labeled Data ---
# Labeled data means the dataset has both:
# 1. Input values (features)
# 2. Correct output/answers (labels)
#
# A "label" is the value we want the model to predict.
# Example:
# Features: [height, weight]
# Label: "male" or "female"
#
# In labeled data:
# Each sample already has the correct answer.
# The model learns by comparing its prediction with the true label.
#
# Supervised learning uses labeled data.
# Example tasks: Classification, Regression


# --- Unlabeled Data ---
# Unlabeled data means the dataset has ONLY:
# 1. Input values (features)
# There is NO correct output/answer given.
#
# Example:
# Features: [height, weight]
# No label like "male/female"
#
# In unlabeled data:
# The model does NOT know the correct answers.
# It tries to find patterns or groups in the data on its own.
#
# Unsupervised learning uses unlabeled data.
# Example tasks: Clustering, Dimensionality Reduction
```

## Types of UnsuperVised Learning
### 1. Clustering
### 2. Association

```python
# --- Clustering ---
# Clustering is an unsupervised learning technique.
# It groups similar data points together based on patterns in the data.
#
# Important point:
# There are NO labels in clustering (unlabeled data).
#
# The algorithm looks at the features and tries to form natural groups.
#
# Example:
# If we have customer data:
#   age, income, spending score
# The algorithm may automatically group customers as:
#   - high spenders
#   - low spenders
#   - medium spenders
#
# Common clustering algorithms:
# - K-Means
# - Hierarchical Clustering
# - DBSCAN
#
# Clustering helps answer:
# "Which data points are similar?"


# --- Association ---
# Association is also an unsupervised learning technique.
# It finds relationships between items or events in a dataset.
#
# Association tries to discover rules like:
# "If a customer buys item A, they also buy item B."
#
# Example:
# In a supermarket dataset:
#   People who buy bread often buy butter.
# This becomes an association rule:
#   bread -> butter
#
# This is used in:
# - Market Basket Analysis
# - Recommendation systems (Amazon, Flipkart)
#
# Common association algorithms:
# - Apriori Algorithm
# - FP-Growth Algorithm
#
# Association helps answer:
# "Which items occur together frequently?"
```
