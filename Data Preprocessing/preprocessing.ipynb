## Ordinal Encoding

```python
# there are two way to do that:
#    1. Cyclic Line
#    2. Map function
```

### 1. Cyclic Line

```python
# first we just make simple dataset for better understanding then we just apply on real or large dataset

df = pd.DataFrame({"Size":['s', 'm','l', 'xl', 's', 'm', 'l', 's', 's', 'l', 'xl', 'm']})
df.head(3)
```

```python
ord_data = [['s', 'm', 'l', 'xl']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = ord_data)
oe.fit(df[["Size"]])
```

```python
df["size_oe"] = oe.transform(df[['Size']])
```

```python
df.head()
```

### 2. Using Map 

```python
ord_data1 = {"s": 0, "m":1, "l":2, "xl":6}
df['size_map'] = df['Size'].map(ord_data1)
df
```

### Remove Outlier using IQR

```python
Q1 = dataset["CoapplicantIncome"].quantile(0.25)
Q3 = dataset["CoapplicantIncome"].quantile(0.75)
```

```python
IQR = Q3 - Q1
```

```python
min_range = Q1 - (1.5 * IQR)
max_range = Q3 + (1.5 * IQR)

print(min_range)
print(max_range)
```

```python
without_otl = df[df['CoapplicantIncome'] <= max_range]
```

```python
sns.boxplot(without_otl.CoapplicantIncome)
```

### Remove Outlier using Z-score

###  = Mean
###  = Standard Deviation

## $$
z = \frac{x - \mu}{\sigma}
$$


```python
# using direct method
import matplotlib.image as mpimg

# read and display image
img = mpimg.imread('z-score.png')
plt.imshow(img)
plt.axis('off')  # hide axes
plt.show()
```

```python
# we are use this 3 for 99.72% range of our data
min_range = df['ApplicantIncome'].mean() - (3 * df['ApplicantIncome'].std())
max_range = df['ApplicantIncome'].mean() + (3 * df['ApplicantIncome'].std())

min_range, max_range
```

```python
new_data = df[df['ApplicantIncome'] <= max_range]
```

```python
# Now that outliers are not harmfull for us
sns.boxplot(new_data['ApplicantIncome'])
```

```python
# now we are doing with z-score
```

```python
z_score = (df['CoapplicantIncome'] - df['CoapplicantIncome'].mean())/df['CoapplicantIncome'].std()
z_score
```

```python
df['z_score'] = z_score
```
