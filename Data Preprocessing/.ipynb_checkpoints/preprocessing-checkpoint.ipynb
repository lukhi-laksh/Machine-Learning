## Ordinal Encoding

```python
# there are two way to do that:
#    1. Cyclic Line
#    2. Map function
```

### 1. Cyclic Line

```python
# first we just make simple dataset for better understanding then we just apply on real or large dataset

df = pd.DataFrame({"Size":['s', 'm','l', 'xl', 's', 'm', 'l', 's', 's', 'l', 'xl', 'm']})
df.head(3)
```

```python
ord_data = [['s', 'm', 'l', 'xl']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = ord_data)
oe.fit(df[["Size"]])
```

```python
df["size_oe"] = oe.transform(df[['Size']])
```

```python
df.head()
```

### 2. Using Map 

```python
ord_data1 = {"s": 0, "m":1, "l":2, "xl":6}
df['size_map'] = df['Size'].map(ord_data1)
df
```

### Remove Outlier using IQR

```python
Q1 = dataset["CoapplicantIncome"].quantile(0.25)
Q3 = dataset["CoapplicantIncome"].quantile(0.75)
```

```python
IQR = Q3 - Q1
```

```python
min_range = Q1 - (1.5 * IQR)
max_range = Q3 + (1.5 * IQR)

print(min_range)
print(max_range)
```

```python
without_otl = df[df['CoapplicantIncome'] <= max_range]
```

```python
sns.boxplot(without_otl.CoapplicantIncome)
```

### Remove Outlier using Z-score

###  = Mean
###  = Standard Deviation

## $$
z = \frac{x - \mu}{\sigma}
$$


```python
# using direct method
import matplotlib.image as mpimg

# read and display image
img = mpimg.imread('z-score.png')
plt.imshow(img)
plt.axis('off')  # hide axes
plt.show()
```

```python
sns.boxplot(without_otl.ApplicantIncome)
```

```python
# we are use this 3 for 99.72% range of our data
min_range = df['ApplicantIncome'].mean() - (3 * df['ApplicantIncome'].std())
max_range = df['ApplicantIncome'].mean() + (3 * df['ApplicantIncome'].std())

min_range, max_range
```

```python
new_data = df[df['ApplicantIncome'] <= max_range]
```

```python
# Now that outliers are not harmfull for us
sns.boxplot(new_data['ApplicantIncome'])
```

```python
# now we are doing with z-score
```

```python
z_score = (df['CoapplicantIncome'] - df['CoapplicantIncome'].mean())/df['CoapplicantIncome'].std()
z_score
```

```python
df['z_score'] = z_score
```

# Feature Scaling

## There are two type of Feature Scaling :
### - 1. Standardization
### - 2. Normalization

#### scalling change effect
#### when we are use this method then our mean become 0 means it's come to center
#### Variance come to -1 to 1
#### we are change value of data not property we are work each row that's why it not show effect

## 1. Standardization

```python
# It is a very effective technique which re-scales a feature value so that it has a distribution with a mean value of 0 and a variance equal to 1.
```

```python
img = mpimg.imread('scalling_change_effect.png')
plt.imshow(img)
plt.axis('off')
plt.show()
```

## $$
X_{\text{new}} = \frac{X_i - \bar{X}_{\text{mean}}}{\text{Standard Deviation}}
$$

```python
df['ApplicantIncome'] = df['ApplicantIncome'].fillna(df['ApplicantIncome'].mean())
```

```python
before = df['ApplicantIncome']
before
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
ss.fit(df[['ApplicantIncome']])
```

```python
df['ApplicantIncome_ss'] = pd.DataFrame(ss.transform(df[['ApplicantIncome']]), columns = ['x'])
```

```python
df.head(3)
```

```python
# Here we can show std**2 = variance around 1 and all data are convert in short form 
df['ApplicantIncome_ss'].describe()
```

```python
# Here also we can check before and now our data's beheve same as before
# Means Nature of Data are same
import warnings
warnings.filterwarnings("ignore")

before = df['ApplicantIncome']
after = df['ApplicantIncome_ss']


plt.subplot(1, 2, 1)
plt.title('Before')
sns.distplot(before)

plt.subplot(1, 2, 2)
plt.title('After')
sns.distplot(after)
plt.show()
```

## Normalization 

```python
# It is a scaling technique in which values are shifted and rescaled so that they end up ranging between 0 and 1. It is also known as Min-Max scaling.
# It's also konwn as Min-Max Scaling
```

## $$
X_{\text{new}} = \frac{X_i - \min(X)}{\max(X) - \min(X)}
$$


```python
# in Normalization all data set x axis and y axis both of there set in 0 to 1
# here below we can see data set to x and y of them 0 to 1
img = mpimg.imread('Normalization.png')
plt.imshow(img)
plt.axis('off')
plt.show()
```
