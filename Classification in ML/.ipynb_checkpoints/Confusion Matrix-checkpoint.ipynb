```python
# A confusion matrix is a simple and useful tool for understanding the performance of a classification model,
#      like one used in machine learning or statistics.

# It helps you evaluate how well your model is doing in categorizing things correctly.

# It is also known as the error matrix.

# The matrix consists of predictions result in a summarized form, 
#     which has number of correct predictions and incorrect predictions.
```

```python
# TN = True Negative
# FN = False Negative
# FP = False Positive
# TP = True Positive
```

```python
# Confusion Matrix Explanation
# ----------------------------
# A confusion matrix is a way to evaluate the performance of a classification model.
# It compares the actual (true) labels with the predicted labels from your model.

# The matrix looks like this:

#                 Predicted
#                 0     1
# True      0    TN    FP
#           1    FN    TP

# Where:
# TN = True Negative: Model correctly predicted class 0 (negative) 
# FP = False Positive: Model predicted 1 (positive) but actual was 0 (negative)
# FN = False Negative: Model predicted 0 (negative) but actual was 1 (positive)
# TP = True Positive: Model correctly predicted class 1 (positive)

# Example using the uploaded confusion matrix:
# -------------------------------------------
# TN = 86  -> 86 samples were actually 0 and predicted as 0
# FP = 7   -> 7 samples were actually 0 but predicted as 1
# FN = 3   -> 3 samples were actually 1 but predicted as 0
# TP = 104 -> 104 samples were actually 1 and predicted as 1

# How data goes into each box:
# -----------------------------
# For each sample in your dataset:
# 1. Check the true label (0 or 1)
# 2. Check the predicted label from the model (0 or 1)
# 3. Place the sample count into the correct box:
#    - If true=0 and predicted=0 -> add to TN
#    - If true=0 and predicted=1 -> add to FP
#    - If true=1 and predicted=0 -> add to FN
#    - If true=1 and predicted=1 -> add to TP

# Python example (conceptual):
# y_true = [0,0,0,1,1,1,1,0,1,0]   # actual labels
# y_pred = [0,1,0,1,0,1,1,0,1,0]   # predicted labels
# TN, FP, FN, TP calculation:
# TN = sum((yt==0 and yp==0) for yt, yp in zip(y_true, y_pred))
# FP = sum((yt==0 and yp==1) for yt, yp in zip(y_true, y_pred))
# FN = sum((yt==1 and yp==0) for yt, yp in zip(y_true, y_pred))
# TP = sum((yt==1 and yp==1) for yt, yp in zip(y_true, y_pred))
# print(TN, FP, FN, TP)

# From the confusion matrix above:
# Accuracy = (TP + TN) / (TP + TN + FP + FN)
```

## Recall (Sensitivity) = TP / (TP + FN)

```python
# It helps us to Measure how many Positive Samples were correctly Classifield by ML Model
```

```python
# if we need Low value in FN then we need to take high our Recall Value
```

## Precision = TP / (TP + FP)

```python
# It helps us to Measure the ability to classify Positive sample in the model.
```

```python
# if we need low value in FP then we need to take high our precision value
```

## F1-Score

```python
# It is the Harmonic mean of Precision and recall. It takes both False Positive and False Negative in Account
# Therefore, it Performs Well on an Imbalanced dataset.
```

$$
F1\ Score = 2 \times \frac{Precision \times Recall}{Precision + Recall}
$$

```python
# we need to focus on Increaes F1 value
# if F1 value is high then model is train best
```

# Imbalanced Data

```python
# Imbalanced data in data science means the number of samples in each class is not equal.
# One class has many more examples, and another class has very few.

# The model becomes biased if there is only few catagory any one
# video ex. Airports on indian People
```

### Example

```python
# Fraud detection  99% normal, 1% fraud
# Medical diagnosis  95% healthy, 5% sick
# Spam detection  90% normal emails, 10% spam
```

## Techniques to handle inbalanced Data
### 1. **Random Under Sampling:** We will reduce the majority of the class so that it will have same no of as the minority
### 2. **Random Over Sampling:** We will increase the size of manority is inactive class to the size of majority class ie active.

## **1. Random Under Sampling**

```python
# Random Under Sampling (RUS)

# Suppose you have imbalanced data:
# Example:
#   900 samples of class 0
#   100 samples of class 1
#
# The dataset is unbalanced because one class (0) is much larger.

# What Random Under Sampling does:
# - It removes (deletes) some data from the majority class (class 0)
# - It reduces the big class down to the size of the smaller class
#
# Example:
#   From 900 samples of class 0  randomly keep only 100 samples
#   Class 1 already has 100 samples  keep all
#
# After RUS:
#   100 samples of class 0
#   100 samples of class 1
#   Balanced dataset!

# Why we use it:
# - To make machine learning models learn equally from both classes
# - To avoid bias toward the majority class

# Downside:
# - We lose some useful data because many samples of the majority class are removed.
```

```python
# we have only 143 numbers of value '1' so imblearn help us to do both value same count
# imblearn (Imbalanced-Learn) Explanation
# ---------------------------------------
# imblearn is a Python library used to handle imbalanced datasets in machine learning.
# In many real-world classification problems, the number of samples in each class is not equal.
# Example: Fraud detection, disease detection, spam detection  where positive cases are very low.
#
# If the dataset is imbalanced, the model may learn to predict only the majority class,
# which gives high accuracy but very poor performance for minority class.
```

## **2. Random Over Sampling**

```python
# Random Over Sampling (ROS)

# Suppose your dataset is imbalanced:
# Example:
#   900 samples of class 0 (majority)
#   100 samples of class 1 (minority)

# What Random Over Sampling does:
# - It increases the number of samples in the minority class (class 1)
# - It does this by randomly repeating (copying) some of the minority samples
#
# Example:
#   Class 1 has 100 samples  randomly duplicate them until it becomes 900
#   Class 0 already has 900  keep as it is

# After ROS:
#   900 samples of class 0
#   900 samples of class 1
#   Balanced dataset!

# Why we use it:
# - To give equal importance to both classes during training
# - It helps the model learn patterns from the minority class

# Downside:
# - Because it duplicates existing data, it can cause overfitting
#   (model learns the same samples again and again)
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ro = RandomOverSampler()
```

```python
ro_x, ro_y = ro.fit_resample(x, y)
```

```python
print(ro_y.value_counts())
```

```python
# In Over Sampling Repete our small amount of data Repete and repeted it to Almost same value
```

```python
x_train, x_test, y_train, y_test = train_test_split(ro_x, ro_y, test_size=0.2, random_state=42)
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
lr.predict([[46, 41000]])
```

```python
# Now it's show answer 1 which actual answer so that mathod also increase accuraccy and give 
# right prediction than before. 
```
