##### • Naive Bayes is a classification algorithm based on Bayes' theorem. 
##### • which is a probability theory that describes the probability of an event, based on prior knowledge of conditions that might be related to the event. 
##### • **Naive** : It is called Naive because it assumes that the occurrence of a certain feature is independent of the occurrence of other features.
##### • **Bayes** : It is called Bayes because it depends on the principle of Bayes' Theorem.

```python
# Probability basics:
# 0 < P(E) < 1
# P(E) = Favourable Outcomes / Total Outcomes

# 1) Independent events:
# Example: Rolling a die → outcomes don’t affect each other.
# So P(rolling 1) = 1/6, P(rolling 5) = 1/6, etc.

# 2) Dependent events:
# Events affect each other (like removing balls from a box).
# Example: Box has 3 Red (R) and 3 Blue (B) → total 6.
# P(R) = 3/6

# If we take 1 Red out, total becomes 5 balls.
# Now probability of Blue depends on Red taken before:
# P(B | R) = 3/5   # (Blue given Red was removed first)
```

```python
# Formula: P(A ∩ B) = P(B | A) * P(A)
# Meaning: Probability of A AND B happening together.

# Also: P(A ∩ B) = P(A | B) * P(B)
# Same value because A ∩ B is the same region in both events.

# Example with sets:
# A = {1, 2, 3, 4}
# B = {4, 5, 6, 7}
# Common part (A ∩ B) = {4}

# So: P(A ∩ B) = 1/8   # only element 4 is common out of 8 total
# And: P(B ∩ A) = 1/8   # same value, order doesn’t change intersection
```

```python
plt.figure(figsize=(5, 4))
img = mpimg.imread('Image/Naive_bays_formula.png')
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# From basic probability:
# P(A ∩ B) = P(B | A) * P(A)
# P(A ∩ B) = P(A | B) * P(B)

# Since both are the same intersection region:
# P(B | A) * P(A) = P(A | B) * P(B)

# Now divide both sides by P(B):
# This gives Bayes' Theorem:

# Bayes Theorem:
# P(A | B) = (P(B | A) * P(A)) / P(B)

# Meaning:
# Probability of A given B = (Likelihood * Prior) / Evidence
```

**Bayes' Theorem :** Bayes' theorem is also known as Bayes' Rule or Bayes' lawwhich is used to determine the probability of a hypothesis with priorknowledge. It depends on the conditional probability.
$$P(A|B) = \frac{P(B|A)P(A)}{P(B)}$$
**Where :** 
##### **$P(A|B)$** is Posterior probability: Probability of hypothesis A on the observed event B.
##### **$P(B|A)$** is Likelihood probability: Probability of the evidence given that the probability a hypothesis is true.
##### **$P(A)$** is Prior Probability: Probability of hypothesis before observing the evidence.
##### **$P(B)$** is Marginal Probability: Probability of Evidence.

```python
# We want the probability of a class Y (yes/no) given features x1, x2, x3.
# Bayes formula for this:
# P(Y | x1, x2, x3) = [ P(x1, x2, x3 | Y) * P(Y) ] / P(x1, x2, x3)

# Meaning:
# - P(Y | features)  → final prediction probability
# - P(features | Y)  → likelihood of seeing x1,x2,x3 if class is Y
# - P(Y)             → prior probability of class Y
# - P(features)      → total probability of x1,x2,x3
```

```python
plt.figure(figsize=(8, 5))
img = mpimg.imread('Image/we_write_it.png')
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# Naive Bayes expands P(Y | x1, x2, x3) assuming all features are independent.
# So joint likelihood becomes a product of individual likelihoods.

# For class "Yes":
# P(Yes | x1, x2, x3) =
#   [ P(Yes) * P(x1 | Yes) * P(x2 | Yes) * P(x3 | Yes) ] 
#   / [ P(x1) * P(x2) * P(x3) ]

# For class "No":
# P(No | x1, x2, x3) =
#   [ P(No) * P(x1 | No) * P(x2 | No) * P(x3 | No) ] 
#   / [ P(x1) * P(x2) * P(x3) ]

# Denominator is the same for all classes,
# so we usually compare only the numerator to pick the best class.
```

```python
plt.figure(figsize=(8, 5))
img = mpimg.imread('Image/belowed_not_count_because_same.png')
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# Naive Bayes calculates probability for each class.
# Example:
# P(Yes | x1, x2, x3) = P(Yes) * P(x1|Yes) * P(x2|Yes) * P(x3|Yes)
# P(No  | x1, x2, x3) = P(No)  * P(x1|No)  * P(x2|No)  * P(x3|No)

# The denominator (P(x1) * P(x2) * P(x3)) is same for both classes.
# So we don't calculate it because it does NOT affect comparison.

# Naive Bayes simply compares:
#     P_yes_score  = P(Yes) * product of likelihoods
#     P_no_score   = P(No)  * product of likelihoods

# Whichever score is higher → that class is the output.
# Example:
# If P_yes_score = 0.75 and P_no_score = 0.25
# Naive Bayes predicts "Yes".
```

## **Three Type of Naive Bayes:**
### 1. Gaussian
### 2. MultiNomial
### 3. Bernouli

## **1. Guassian:**
### Assumes that continuous features follow a Guassian (normal) distribution
### Suitable for features that are continuous and have a normal distribution

## **2. Bernouli:**
### Assumes that features are binary(Boolean) Variables.
### Suitable for data that can be represented as binary features, such as document classification problems where each term is eighter present or absent

## **Multinomial Naive Bayes:**
### Assumes that features follow a multinomial distribution.
### Typically used for discreate data, such as text data, where each feature represents the frequency of a team.

```python
df = pd.read_csv('../Dataset/placement_.csv')
df
```

```python
sns.scatterplot(x = 'cgpa', y='resume_score', data=df, hue='placed')
plt.show()
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1:]
x.head()
```

```python
# we can show we have Almost Normal Distrubution Curv
sns.kdeplot(data=df['cgpa'])
plt.show()
```

```python
# we can show we have best Normal Distrubution Curv
# so we choose 
sns.kdeplot(data=df['resume_score'])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
mln = MultinomialNB()
mln.fit(x_train, y_train)
```

```python
ber = BernoulliNB()
ber.fit(x_train, y_train)
```

```python
print('Score of GussianNB is:', gnb.score(x_test, y_test)*100)
print('Score of MultinomialNB is:', mln.score(x_test, y_test)*100)
print('Score of BernoulliNB is:', ber.score(x_test, y_test)*100)
print('result: so we can say GussianNB is best for us')
```
